## LDA Topic Modeling

```python
import pandas as pd
data = pd.read_csv('data_cleaned_V4.csv',engine='python', dtype=str)
corpus=data['comment'].tolist()
```

```python
import gensim
from gensim import corpora
from nltk.tokenize import word_tokenize

corpsplit=[word_tokenize(doc) for doc in data['comment']]
dictionary = corpora.Dictionary(corpsplit)
doc_term_matrix = [dictionary.doc2bow(doc) for doc in corpsplit]
```

```python
Lda = gensim.models.ldamodel.LdaModel
```

#### num_topics=5

```python
ldamodel = Lda(doc_term_matrix, num_topics=5, id2word = dictionary, passes=50)
```

```python
print(ldamodel.print_topics(num_topics=5, num_words=5))
```

#### num_topics=10

```python
ldamodel2 = Lda(doc_term_matrix, num_topics=10, id2word = dictionary, passes=50)
```

```python
print(ldamodel2.print_topics(num_topics=10, num_words=10))
```

#### num_topics=20

```python
ldamodel3 = Lda(doc_term_matrix, num_topics=20, id2word = dictionary, passes=50)
```

```python
print(ldamodel3.print_topics(num_topics=20, num_words=10))
```

## Word Frequency

```python
import nltk
```

```python
wordfreq = {}
for sentence in corpus:
    tokens = sentence.split()
    for token in tokens:
        if token not in wordfreq.keys():
            wordfreq[token] = 1
        else:
            wordfreq[token] += 1
```

```python
wordfreq['cannot']
```

```python
print(wordfreq)
```

## Bag of Words

```python
import heapq
most_freq = heapq.nlargest(200, wordfreq, key=wordfreq.get)
```

```python
sentence_vectors = []
for sentence in corpus:
    sentence_tokens = nltk.word_tokenize(sentence)
    sent_vec = []
    for token in most_freq:
        if token in sentence_tokens:
            sent_vec.append(1)
        else:
            sent_vec.append(0)
    sentence_vectors.append(sent_vec)
```

```python
print(sentence_vectors)
```

#### Matrix Form

```python
import numpy as np
sentence_vectors_Matrix = np.asarray(sentence_vectors)
```

```python
print(sentence_vectors_Matrix)
```

## Ngrams

#### Unigrams

```python
import re
import collections
c = collections.Counter(wordfreq)
c.most_common(100)
```

#### Bigrams

```python
from nltk import word_tokenize
from nltk.util import ngrams
from nltk import FreqDist

bigramfdist = FreqDist()
for text in corpus:
    if(len(text)>1):
        token = nltk.word_tokenize(text)
        bigrams = ngrams(token,2)
        bigramfdist.update(bigrams)
```

```python
list(bigramfdist)
```

```python
bigramfdist.most_common(10)
```

#### Trigrams

```python
threeramfdist = FreqDist()

for text in corpus:
    if(len(text)>1):
        token = nltk.word_tokenize(text)
        if len(token)>2:
            trigrams = ngrams(token,3)
            threeramfdist.update(trigrams)
```

```python
list(threeramfdist)
```

```python
threeramfdist.most_common(10)
```

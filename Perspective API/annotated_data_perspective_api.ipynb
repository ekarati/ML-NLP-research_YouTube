##  Using Perspective API to score Comments

```python
import json 
import requests 
import pandas as pd
import time

api_key = 'hidden_api_key'
url = ('https://commentanalyzer.googleapis.com/v1alpha1/comments:analyze' +    
    '?key=' + api_key)
```

```python
data = pd.read_csv('data_cleaned_V2.csv',engine='python', dtype=str)
df = pd.DataFrame(data, columns= ['comment'])
```

```python
ls=[]
for i in df['comment']:
    data_dict = {
                'comment': {'text': i},
                'languages': ['en'],
                'requestedAttributes': {'TOXICITY': {}, 'SEVERE_TOXICITY': {}, 'IDENTITY_ATTACK': {}, 'INSULT': {}, 
                                        'PROFANITY': {}, 'THREAT': {},'SEXUALLY_EXPLICIT': {},'FLIRTATION': {},
                                        'ATTACK_ON_AUTHOR': {},'ATTACK_ON_COMMENTER': {},'INCOHERENT': {},'INFLAMMATORY': {},
                                        'LIKELY_TO_REJECT': {},'OBSCENE': {},'SPAM': {},'UNSUBSTANTIAL': {}}
                }

    response = requests.post(url=url, data=json.dumps(data_dict)) 
    rdict = json.loads(response.content)

    T=rdict['attributeScores']['TOXICITY']['summaryScore']['value']
    ST=rdict['attributeScores']['SEVERE_TOXICITY']['summaryScore']['value']
    IA=rdict['attributeScores']['IDENTITY_ATTACK']['summaryScore']['value']
    IN=rdict['attributeScores']['INSULT']['summaryScore']['value']
    P=rdict['attributeScores']['PROFANITY']['summaryScore']['value']
    TR=rdict['attributeScores']['THREAT']['summaryScore']['value']
    SE=rdict['attributeScores']['SEXUALLY_EXPLICIT']['summaryScore']['value']
    F=rdict['attributeScores']['FLIRTATION']['summaryScore']['value']
    AA=rdict['attributeScores']['ATTACK_ON_AUTHOR']['summaryScore']['value']
    AC=rdict['attributeScores']['ATTACK_ON_COMMENTER']['summaryScore']['value']
    IC=rdict['attributeScores']['INCOHERENT']['summaryScore']['value']
    IF=rdict['attributeScores']['INFLAMMATORY']['summaryScore']['value']
    LR=rdict['attributeScores']['LIKELY_TO_REJECT']['summaryScore']['value']
    O=rdict['attributeScores']['OBSCENE']['summaryScore']['value']
    S=rdict['attributeScores']['SPAM']['summaryScore']['value']
    U=rdict['attributeScores']['UNSUBSTANTIAL']['summaryScore']['value']
    
    print(T, ST, IA, IN, P, TR, SE, F, AA, AC, IC, IF, LR, O, S, U, i)
    ls.append((T, ST, IA, IN, P, TR, SE, F, AA, AC, IC, IF, LR, O, S, U, i))
    time.sleep(1.5)
```

```python
ndf = pd.DataFrame(ls, columns=('TOXICITY', 'SEVERE_TOXICITY', 'IDENTITY_ATTACK', 'INSULT', 
                                'PROFANITY', 'THREAT','SEXUALLY_EXPLICIT','FLIRTATION',
                                'ATTACK_ON_AUTHOR','ATTACK_ON_COMMENTER','INCOHERENT','INFLAMMATORY',
                                'LIKELY_TO_REJECT','OBSCENE','SPAM','UNSUBSTANTIAL', 'COMMENT'))
ndf.to_csv('data_perspective_api_scores.csv')
```
